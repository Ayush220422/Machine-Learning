```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
data =pd.read_csv('student_placements.csv')
```

```python
type(data)
```

```python
data.tail()
```

```python
data.drop('Student',axis=1,inplace=True)
```

```python
data.head(36)
```

```python
plt.scatter(data['CGPA'],data['Placement_Package_LPA'])
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
x= data.iloc[:,0:1]
y=data.iloc[:,[-1]]
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test.iloc[0].values.reshape(1,1))
```

```python
plt.scatter(data['CGPA'],data['Placement_Package_LPA'])
plt.plot(x_train,lr.predict(x_train),color='red')
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
lr.coef_
```

```python
plt.scatter(x, y)
plt.plot(x, lr.predict(x), color='red')
# plt.xlim(0,10)
# plt.ylim(0,20)
plt.axhline(lr.intercept_[0], color='green', linestyle='--')  # intercept line
plt.show()
```
